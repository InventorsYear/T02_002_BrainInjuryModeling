### 5.1 Newton's 2nd Law

Consider motion that is only in one dimension, say initially along the horizontal x-axis (e.g. along a track on the lab table).

Newton's Second Law states that the net force on an object is equal to the mass of the object times its acceleration.

$$ \begin{equation}
\sum_{i}{{F}_i} = m{a}
\end{equation} $$

Let's state this another way: the acceleration $a$ is proportional to the net force with proportionality constant $\frac{1}{m}$. 

$$ \begin{equation}
a = \frac{1}{m}\sum_{i}{{F}_i}
\end{equation} $$

In this context, we say that $m$ measures the *inertia* of the object: the larger the value of m and hence the larger the inertia, the smaller the object's response to a net force.

### 5.2 Kinematics of accelerated motion in one dimension

If the net force and mass remain constant, then the acceleration $a$ is constant. Now the definition of acceleration is:

$$\begin{equation}
\frac{d^2x}{dt^2}\equiv a
\end{equation}$$

For constant $a$, then we can integrate twice to find $x$ as a function of time t given an initial velocity $v_0$ and initial position $x_0$:

$$ \begin{equation}\begin{aligned}
\frac{d^2x}{dt^2}&=a \\
v &\equiv \frac{dx}{dt} = v_0 + \int_{0}^{t}\frac{d^2x}{dt^2}dt \:=\: v_0 + \int_{0}^{t}adt \:=\: v_0 + at\\
x &= x_0 + \int_{0}^{t}\frac{dx}{dt}dt \:=\: x_0 + \int_{0}^{t}(v_0 + at)adt \:=\: x_0 + v_0t + \frac{1}{2}at^2
\end{aligned}\end{equation}$$

### 5.3 Motion in the vertical direction
If we now turn our axes and work in the vertical direction, we will use the variable $y$ to represent the position of the mass. Let's also assume that gravity is acting downwards in the $-y$ direction. For an object of mass $m$, the force of gravity is found empirically to be proportional to the mass:

$$\begin{equation}
F_\text{grav}=-gm
\end{equation}$$

where the proportionality constant $g$ = 9.80 N/Kg and the negative sign indicates that the force is acting downwards.

If gravity is the only force acting on an object, we say the object is in free fall. Here we neglect other realistic forces like air resistance, also known as fluid drag, a form of friction. For our laboratory model of head impact explored by doing "drop tests", this is a reasonable assumption - at least initially. Different thinking might apply, for example, in dealing with falls from larger heights ... including situations  important in protecting small delicate equipment from large falls.

So for free fall under gravity:

$$ \begin{equation}\begin{aligned}
F_\text{grav}&=ma \\
-gm &= ma\\
-g &= a
\end{aligned}\end{equation}$$

We have arrived at the fact that the acceleration is independent of mass, 

$$\begin{equation}
a = -g
\end{equation}$$

where we can state the value of $g$ in equivalent units $g$ = 9.80 $\frac{\text{m}}{\text{s}^2}$.

Thus the motion under gravitational force alone is described kinematically by:

$$ \begin{equation}\begin{aligned}
\frac{d^2y}{dt^2}&=-g \\
v &\equiv \frac{dy}{dt} = v_0 - gt\\
y &= y_0 + v_0t - \frac{1}{2}gt^2
\end{aligned}\end{equation}$$

The companion relation for the speed is:

$$\begin{equation}
v^2-v_0^2=-2g(y-y_0).
\end{equation}$$

### 9.1 Free fall


![Bottle in free fall](Figures/CokeBottleWithHangingBatteryFreeFall.png)

*Figure. System under free fall. The stretched-paper assembly is dropped from an initial height h and accelerates downwards at the rate g.*

Here is an analysis of what is going on. Consider first the free fall situation.

What is the tension force $T$ in the folded paper towel as it is in free fall? You might say it is mg but that’s not correct. Here’s why: Draw a free-body diagram and conclude that

$$ m  a_\text{freefall} = T_\text{free_fall}  –  m g.$$

Since $a_\text{freefall} = – g$, we conclude that $T_\text{free_fall} = 0$.


### 9.2 Collision force and acceleration

![Bottle after impact](Figures/afterImpactReduced.jpg)
*Figure. The system has been dropped onto the floor from a height sufficient to cause the folded paper to tear and the battery to drop to the bottom of the bottle. 

![Colliding bottle](Figures/CokeBottleWithHangingBatteryCollision.png)

*Figure. The system suddenly impacts the floor. A large acceleration occurs as the system changes its velocity from the final falling velocity to zero in a short amount of time.*

At the bottom of the fall, the system collides with the floor. The collision acceleration can be related to the forces by Newton’s second law:

$$ma_\text{collision} = T_\text{collision} – mg.$$

Thus 

$$T_\text{collision} = m(a_\text{collision} + g).$$

We can predict the collision acceleration from the mass of bottle plus water that just causes the folded paper towel to break. Suppose we find that the folded paper towel breaks when $m_{\text{b+w}} \equiv m_\text{static_break}$. Now we make a key assumption in our modeling: the breaking force is the same in both the static and collision cases. Later we might want to revise this assumption and refine our model, but for now we predict that the folded paper towel breaks upon collision when:

$$T_\text{collision} = T_\text{static_break}.$$

Thus:

$$m(a_\text{collision} + g) = m_\text{static_break}g.$$

Solving for $a_\text{collision}$:

$$a_\text{collision} = \frac{m_\text{static_break} – m}{ m}g.$$

Algebra gives an equivalent expression:

$$a_\text{collision} = \left(\frac{m_\text{static_break}}{m} – 1\right) g.$$

The larger the ratio $m_\text{static_break}/m$, the larger $a_\text{collision}$, that is the larger the number of “g’s” the collision must create in order to tear the folded paper towel.

As a check on the result, note that $a_\text{collision} = 0$ when $m = m_\text{static_break}$, that is one does not even need to even drop the system for the folded paper towel to break. This is exactly the static situation.

### 9.3 Collision Time

If the system as fallen a distance h, we can calculate the velocity just before impact from the kinematic relation descibing the change in velocity when an object accelerates at constant rate $g$ through a distance $h$:

$$ v_\text{collision}^2 – v_0^2 = 2gh. $$

Since the initial speed $v_0$ is zero, the velocity at the point of collision

$$ v_\text{collision} = – \sqrt{2gh}.$$

The collision velocity is negative because it is downward. 

Observe a few trial collisions, perhaps without the folded paper towel and battery, to see what happens during and after the collsion. Probably you'll see the "skull" (the bottle and lid) strike the ground and bounce a short distance upwards a few times in quick succession. We'll make a simplifying assumption that the bouncing rebound is negligable: that is, that the kinetic energy of the fall is almost entirely absorbed in the collision with little left over for rebound motion. So we'll take the velocity after collision to be nearly zero and estimate that the change in velocity during the collision is given by

$$ \Delta v = v_\text{rebound} - v_\text{collision} \approx 0 - v_\text{collision}.$$

$$ \Delta v \approx  - ( – \sqrt{2gh})$$

$$ \Delta v \approx \sqrt{2gh}.$$

We relate this to the acceleration by:

$$ a_\text{collision} = \frac{\Delta v}{\Delta t}.$$

This is actually the acceleration averaged over the short time interval between the start of the collision and the time when the system is totally at rest. Our model for now will assume that this average acceleration is well approximated by the acceleration defined by the breaking strength of the folded paper towel. With more advanced instrumentation such as an accelerometer and a method for fast data recording, we can later test this assumption and refine our model.

For now, we want to estimate $\Delta t$, the unkown short time of the collision with the hard surface. Rearranging the above expression for the collision acceleration:

$$ \Delta t = \frac{\Delta v}{a_\text{collision}}.$$

From now on, we'll assume that we are dropping from the critical height $h=h_c$ where the folded paper towel just breaks. Inserting expressions for $\Delta v$ and $a_\text{collision}$, we get

$$ \Delta t = \frac{\sqrt{2gh_c}}{\left(\frac{m_\text{static_break}}{m} – 1\right) g}.$$

$$ \Delta t = \frac{1}{\left(\frac{m_\text{static_break}}{m} – 1\right)}\sqrt{\frac{2h_c}{g}}.$$

This is a key result that we will use for further analysis below. However, we can  also give this an interesting interpretation. The time $t_\text{fall}$ for an object to fall a distance h starting at rest satisfies the kinematic relation:

$$h = \frac{1}{2}gt_\text{fall}^2.$$

So 

$$ t_\text{fall_c} = \sqrt{\frac{2h_c}{g}}.$$

We see then that the collision time is shortened by the same factor that increased the number of "g's" of the collision acceleration:

$$ \Delta t = \frac{1}{\left(\frac{m_\text{static_break}}{m} – 1\right)}t_\text{fall_c}.$$

#### 9.3.1 Tissue parameter estimation

The collision time is appropriate to our particular circumtances, but does not tell us all that we need to know if we want to design protection equipment. It does at least give us a clue that in order to prevent damage from this particular drop height, we need to lengthen the collision time - for example by placing some soft padding where the assembly hits the ground. But how can we extend this to larger heights? How do we know how much padding is needed

What we really want to know is a parameter of the "brain tissue" that allows us to think about our design over a wider range of circumstances: different heights of dropping, or different speeds in - say - a head-on collision with a surface such as hitting a tree while skiing.

To do this, we're going to add an important element to our model. We will assume that the brain is suspended from the skull by a spring. The spring increases its force the more it is stretched. Since in this configuration the stretching occurs in the vertical direction, we'll use $\Delta y$ to denote the amount of stretching.

![Srping model of tension in the folded paper](Figures/CokeBottleWithHangingBatterySpringModel.png)

*Figure. When the system impacts the floor, the "brain mass" (the battery) deflect downwards a distance $\Delta y$, stretching the supporting folded paper towel as though it were a spring with spring constant k.*

The force developed by the spring grows with stretching:

$$F_\text{spring} = k \Delta y $$

(Here we are only interested in the magnitude of the force. Later we might pay more attention to the sign, i.e. to the direction.)

Finally, when this force exceeds the breaking strength, the supporting tissue ruptures. The situation

$$F_\text{spring} > T_\text{static_break}$$ 

implies rupture of the folded paper towel!

In order to connect this with the collision time $\Delta t$, this is we will take a short cut and use energy concepts. If you have not yet encountered these ideas in lecture courses, don't worry! We'll use these now to arrive at a prediction and - like good scientists - let experiment help us see if our reasoning is on the right track.

Just before the "skull-and-tissue" hits the ground, the brain has a form of energy called kinetic energy:

$$E_\text{kinetic} = \frac{1}{2}mv_\text{collision}^2$$

When the skull-and-tissue hits the ground, we will assume that  kinetic energy of the brain is converted into a form called "potential energy" when stretching the spring:

$$E_\text{potential} = \frac{1}{2}k(\Delta y)^2$$

We are using an idea called conservation energy: when the brain comes to rest, all of its kinetic energy is converted into potential energy.

$$E_\text{kinetic} = E_\text{potential}$$

$$\frac{1}{2}mv_\text{collision}^2 = \frac{1}{2}k(\Delta y)^2$$

This gives an expression for the amount of deformation caused to bring the brain to rest:

$$\Delta y = \sqrt{\frac{m}{k}}{v_\text{collision}}.$$

We can use this to make an independent estimate of our collision time! First, note that the average velocity during the time interval between the start of the collision and when the "brain" comes to rest is

$$v_\text{av} \equiv \frac{\Delta y}{\Delta t}.$$

If we make a rough assumption that the collision acceleration is constant, then the average velocity is also related to the initial collision velocity by:

$$v_\text{av} = v_\text{collision}/2$$

Thus
$$\Delta t = \frac{\Delta y}{v_\text{av}} = \frac{\Delta y}{v_\text{collision}/2}=\frac{2\Delta y}{v_\text{collision}}$$

Rearranging the expression above for the deformation $\Delta y$, 

$$\frac{\Delta y}{v_\text{collision}}=\sqrt{\frac{m}{k}}$$

Substitute this into the new relation for the collision time to get: 

$$\Delta t = 2\sqrt{\frac{m}{k}}$$.

Amazing deduction! Now finally we can arrive at an estimate of the important brain modeling parameter $k$ by recalling our earlier result for $\Delta t$:

$$ \Delta t = \frac{1}{\left(\frac{m_\text{static_break}}{m} – 1\right)}\sqrt{\frac{2h_c}{g}}.$$

We have two independent expressions for $\Delta t$, so let's equate the right-hand sides of these expressions to one another:

$$\frac{1}{\left(\frac{m_\text{static_break}}{m} – 1\right)}\sqrt{\frac{2h_c}{g}}=2\sqrt{\frac{m}{k}}.$$

Solving for k:

$$ k = 2\frac{mg}{h_c}\left(\frac{m_\text{static_break}}{m} – 1\right)^2. $$

Note: a more careful analysis of the actual non-constant force and non-constant acceleration of the mass attached to the spring gives a result a factor of 4 times smaller (with a small correction also in the mass ratio term):

$$ k = \frac{1}{2}\frac{mg}{h_c}\left(\frac{m_\text{static_break}}{m}\right)^2\left(1-\frac{m}{m_\text{static_break}}\right). $$

(This analysis will be added to the appendix. It requires understanding of the motion of mass-spring systems that involves solving a second-order differential equation that is usually taught later in the first semester of an introductory physics course.)

## 10 Further explorations

### 10.1 Softening the collision

Next drop the assembly from the previously determined critical height onto a piece of soft foam and observe that the towel does not break. Think about how to explain the difference between collision with a hard surface and collision with a pillow.

What's going on? What does it mean for some material to "soften" the collison?

An approach is to consider the soft padding as a second soft spring in series with the original spring representing the stretched tissue. When two springs with spring constants $k_1$ and $k_2$ are in series, their net spring constant k is given by:

$$ \frac{1}{k}=\frac{1}{k_1}+\frac{1}{k_2} $$.
